```python
import pymorphy2
from nltk.corpus import stopwords
import nltk
nltk.download('stopwords')
russian_stopwords = stopwords.words("russian")
```

```python
def print_not_alpha(text):
  
  punct = []
  for chap in text:
    for i in chap:
      if i.isalpha() == False and i != ' ':
        punct.append(i)
  return list(set(punct))
```

```python
puntct = print_not_alpha(text)
```

```python
def delete_my_punct(text, punct = puntct):
  for i in punct:
    for k in range(len(text)):
      try:
        text[k] = text[k].replace(i, '')
      except Exception:
        pass
  return text
```

```python
def delete_spaces(text):
  for i in range(len(text)):
    text[i] = text[i].split(' ')
    text[i] = [k for k in text[i] if k]
  return text
```

```python
text = delete_my_punct(text)
```

```python
text = delete_spaces(text)
```

```python
def delete_stop_words(text):
  for i in range(len(text)):
    text[i] = [word for word in text[i] if word not in russian_stopwords]
  return text
```

```python
text = delete_stop_words(text)
```

```python
m = pymorphy2.MorphAnalyzer()

def normalization(text):
  for i in range(len(text)):
    text[i] = [m.parse(word)[0].normal_form for word in text[i]]
    print(i)
  return text
```

```python
text = normalization(text)
```
